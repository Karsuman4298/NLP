```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("khulasasndh/game-of-thrones-books")

print("Path to dataset files:", path)
```

```python
import os
```

```python
from nltk import sent_tokenize
from gensim.utils import simple_preprocess
import nltk
nltk.download('punkt')
nltk.download('punkt_tab') # Added to address the specific LookupError

story=[]
for filename in os.listdir(path):
  # Attempt to open with 'latin-1' encoding
  with open(os.path.join(path,filename), encoding='latin-1') as f:
    corpus=f.read()
  raw_sent=sent_tokenize(corpus)
  for sent in raw_sent:
    story.append(simple_preprocess(sent))
```

```python
len(story)
```

```python
story[0]
```

```python
model=word2vec.Word2Vec(
    window=10,
    min_count=2,
    vector_size=100,

)
```

```python
model.build_vocab(story)
```

```python
model.train(story,total_examples=model.corpus_count,epochs=model.epochs)
```

```python
model.wv.most_similar("king")
```

```python
model.wv.doesnt_match(['jon','arya','sansa','bran','robb'])
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=3)
X=pca.fit_transform(model.wv.get_normed_vectors())
```
