# Removal of HTML Tags


```python
def tag_removal(text):
    pattern=re.compile('<.*?>')
    return pattern.sub(r'',text)

```

```python
clean_text=tag_removal(df['review'][0])
clean_text
```

```python
df["review"] = df["review"].apply(tag_removal)
```

# Removing punctuation

```python
import string
exclude= string.punctuation
exclude
```

```python
def remove_punc(text):
  for char in exclude:
    text=text.replace(char,'')
  text=re.sub('[0-9]+','',text)
  return text
```

```python
df["review"][1]
```

```python
remove_punc(df["review"][1])
```

```python
import time

start=time.time()
df["review"]=df["review"].apply(remove_punc)
end=time.time()
print(end-start)
```

```python
import nltk
nltk.download('stopwords')
from nltk.corpus import stopwords

stopwords.words('english')
```

```python
def remove_stopword(text):
  new_text=[]


  for words in text.split():
    if words in stopwords.words('english'):
      new_text.append('')
    else:
      new_text.append(words)


  x=new_text[:]
  new_text.clear()
  return " ".join(x)
```

```python
remove_stopword(df['review'][0])
```

# Stemming

```python
from nltk.stem.porter import PorterStemmer
```

```python
ps=PorterStemmer()

def stem_words(text):
  return " ".join([ps.stem(word) for word in text.split()])
```

```python
sample="like liking liked"
```

```python
stem_words(sample)
```

# Lemmatization

```python
from nltk.stem import WordNetLemmatizer
wl=WordNetLemmatizer()
```

```python
sentence="One of the other reviewers has mentioned that after watching just 1 Oz episode you'll be hooked."
```

```python
import nltk
nltk.download('wordnet')

punc = string.punctuation
sentence_token = word_tokenize(sentence)
cleaned_sentence_token = [word for word in sentence_token if word not in punc]


print("{0:20}{1:20}".format('Word','Lemma'))
for words in cleaned_sentence_token:
  print("{0:20}{1:20}".format(words, wl.lemmatize(words, pos='v')))# verb
```
